# logging function

```python
from datetime import datetime
import ipyparams

def file_name():
    currentNotebook = ipyparams.notebook_name
    split_name = currentNotebook.split(".")
    return str(split_name[0])

def log(prioity,messenage):
    #determine file name
    log_file = file_name() + "_log_file.txt"
    print(log_file)
    #logging function
    datetime_stamp = "%Y-%m-%d %H:%M:%S"
    now_time = datetime.now()
    datetime_log = now_time.strftime(datetime_stamp)
    with open(log_file,'a') as f:
        f.write(datetime_log + ',' + '[' + prioity + ']' + messenage + '\n') 
```

```python
log("action","log in again")
```

# Creating a stored procedure

create or replace procedure prac_transfer(
   sender int,
   receiver int, 
   amount dec
)
language plpgsql    
as $$
begin
    -- subtracting the amount from the sender's account 
    update accounts 
    set balance = balance - amount 
    where id = sender;

    -- adding the amount to the receiver's account
    update accounts 
    set balance = balance + amount 
    where id = receiver;

    commit;
end;$$;

call prac_transfer(1, 2, 1000);

```python
%%sql
    CREATE OR REPLACE PROCEDURE  city_index.UPDATE_LEADERS_SCORE(
        in_school_id int, in_leader_score int)
    LANGUAGE plpgsql
    AS
    $$
    BEGIN
        UPDATE city_index.public_schools
        SET leaders_score = in_leader_score
        WHERE school_id = in_school_id;
        
        IF not found THEN
            RAISE EXCEPTION 'Not found the School_id';
        ELSE
            IF in_leader_score < 100 AND in_leader_score > 79 THEN
                UPDATE city_index.public_schools
                SET leaders_icon = 'Very strong'
                WHERE school_id = in_school_id;
            ELSEIF in_leader_score < 80 AND in_leader_score > 59 THEN
                UPDATE city_index.public_schools
                SET leaders_icon = 'Strong'
                WHERE school_id = in_school_id;
            ELSEIF in_leader_score < 60 AND in_leader_score > 39 THEN
                UPDATE city_index.public_schools
                SET leaders_icon = 'Average'
                WHERE school_id = in_school_id;
            ELSEIF in_leader_score < 40 AND in_leader_score > 19 THEN
                UPDATE city_index.public_schools
                SET leaders_icon = 'Weak'
                WHERE school_id = in_school_id;
            ELSEIF in_leader_score < 20 AND in_leader_score >= 0 THEN
                UPDATE city_index.public_schools
                SET leaders_icon = 'Very weak'
                WHERE school_id = in_school_id;
            ELSE
                RAISE EXCEPTION 'out of range';
            END IF;
        END IF;

        commit;
    END;
    $$;
    
```

```python
%%sql
    CALL city_index.UPDATE_LEADERS_SCORE(610038,38);
    CALL city_index.UPDATE_LEADERS_SCORE(610281,101);
```
