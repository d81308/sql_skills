# Optimize with Postgresql
> change data type, then optimize

> Vaccum statement

### view data information

view tables in schema

```python
%%sql
    select *
    from pg_catalog.pg_tables
    where schemaname = 'bookings'
```

target:aircrafts_data table

```python
%%sql
    SELECT *
    FROM bookings.aircrafts_data
```

view data type

```python
%%sql
    SELECT *
    FROM information_schema.columns
    WHERE table_schema = 'bookings'
    AND table_name = 'aircrafts_data'
    ;
```

### optimize step

drop view (if exists in this job)

```python
%%sql
    DROP VIEW bookings.aircrafts;
```

change int into smallint

```python
%%sql
    ALTER TABLE bookings.aircrafts_data ALTER COLUMN range TYPE smallint;
```

check data type again

```python
%%sql
    SELECT *
    FROM information_schema.columns
    WHERE table_schema = 'bookings'
    AND table_name = 'aircrafts_data'
    ;
```
