# logging function

```python
from datetime import datetime
import ipyparams

def file_name():
    currentNotebook = ipyparams.notebook_name
    split_name = currentNotebook.split(".")
    return str(split_name[0])

def log(prioity,messenage):
    #determine file name
    log_file = file_name() + "_log_file.txt"
    print(log_file)
    #logging function
    datetime_stamp = "%Y-%m-%d %H:%M:%S"
    now_time = datetime.now()
    datetime_log = now_time.strftime(datetime_stamp)
    with open(log_file,'a') as f:
        f.write(datetime_log + ',' + '[' + prioity + ']' + messenage + '\n')
```

```python
log("action","ready to start query")
```

### read csv into dataframe

```python
def choose_csv_file(index):
    ##choose csv file, naming is too long, prevent key-in failure
    all_csv_file = glob.glob('download/*.csv')
    print(all_csv_file)
    print("list of all csv files:\n",all_csv_file)

    #choose target file
    import_csv_file = all_csv_file[0]
    
    return import_csv_file
```

### create table into database, server choose sqlite3

```python
##build db and connect to db
#prevent key-in failure, make variable for db name
db_name = "RealWorldData.db"

#sqlite3 connection
con = sqlite3.connect(db_name)
cur = con.cursor()

#magic connection
%sql sqlite:///{db_name}
```
